```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from skimage.feature import hog
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.neural_network import MLPClassifier

# -----------------------------
# Part 1: Data Preprocessing
# -----------------------------

# Paths
DATA_DIR = 'data'
OUTPUT_DIR = 'segmented_signatures'

# Create output directory if it doesn't exist
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

def preprocess_image(image_path):
    """
    Preprocess the image for contour detection.
    """
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Apply Gaussian Blur
    blurred = cv2.GaussianBlur(gray, (5,5), 0)
    # Apply adaptive thresholding
    thresh = cv2.adaptiveThreshold(blurred, 255, 
                                   cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                   cv2.THRESH_BINARY_INV, 11, 2)
    return image, thresh

def find_contours(thresh):
    """
    Find contours in the thresholded image.
    """
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    return contours

def extract_cells(image, thresh):
    """
    Extract cells from the table by finding contours.
    """
    contours = find_contours(thresh)
    # Assuming that cells are rectangular and have a certain size
    cells = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        # Filter out small contours that are unlikely to be cells
        if w > 50 and h > 30:
            cells.append((x, y, w, h))
    # Sort cells top to bottom, then left to right
    cells = sorted(cells, key=lambda b: (b[1], b[0]))
    return cells

def segment_image(image_path):
    """
    Segment the image into cells.
    """
    image, thresh = preprocess_image(image_path)
    cells = extract_cells(image, thresh)
    # Assuming fixed number of columns (5)
    num_cols = 5
    rows = []
    current_row = []
    last_y = -1
    tolerance = 10  # Tolerance for y-coordinate to group cells into the same row
    for cell in cells:
        x, y, w, h = cell
        if last_y == -1:
            last_y = y
        if abs(y - last_y) > tolerance:
            rows.append(current_row)
            current_row = []
            last_y = y
        current_row.append(cell)
    if current_row:
        rows.append(current_row)
    # Sort each row's cells by x-coordinate
    for row in rows:
        row.sort(key=lambda b: b[0])
    return rows, image

def save_signatures(rows, image_name):
    """
    Save signature images into corresponding ID folders.
    """
    for row in rows:
        if len(row) != 5:
            continue  # Skip incomplete rows
        id_cell = row[0]
        signature_cells = row[1:]
        # Extract ID
        x, y, w, h = id_cell
        id_image = original_image[y:y+h, x:x+w]
        id_gray = cv2.cvtColor(id_image, cv2.COLOR_BGR2GRAY)
        _, id_thresh = cv2.threshold(id_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        # Assuming ID is in text format, use OCR or simple binary conversion
        # For simplicity, let's assume the ID is the image name's number
        # Alternatively, you can implement OCR here
        id_text = image_name.split('.')[0]  # e.g., '1' from '1.jpg'
        id_folder = os.path.join(OUTPUT_DIR, id_text)
        if not os.path.exists(id_folder):
            os.makedirs(id_folder)
        # Save signature images
        for idx, sig_cell in enumerate(signature_cells):
            sx, sy, sw, sh = sig_cell
            signature = original_image[sy:sy+sh, sx:sx+sw]
            # Resize signature to a fixed size (e.g., 128x128)
            signature_resized = cv2.resize(signature, (128, 128))
            sig_filename = f"{id_text}_sig{idx+1}.png"
            sig_path = os.path.join(id_folder, sig_filename)
            cv2.imwrite(sig_path, signature_resized)

# Process each image in the data directory
for image_name in os.listdir(DATA_DIR):
    if image_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        image_path = os.path.join(DATA_DIR, image_name)
        rows, original_image = segment_image(image_path)
        save_signatures(rows, image_name)
        print(f"Processed {image_name}")

# -----------------------------
# Part 2: Model Selection
# -----------------------------

# Parameters
IMG_HEIGHT, IMG_WIDTH = 128, 128
BATCH_SIZE = 32
EPOCHS = 50
NUM_CLASSES = len(os.listdir(OUTPUT_DIR))  # Dynamically set number of classes

# Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,  # 80-20 split
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Training Generator
train_generator = train_datagen.flow_from_directory(
    OUTPUT_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

# Validation Generator
validation_generator = train_datagen.flow_from_directory(
    OUTPUT_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

# Building the CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    MaxPooling2D(2,2),
    
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(NUM_CLASSES, activation='softmax')
])

# Compile the model
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()

# -----------------------------
# Part 3: Model Evaluation
# -----------------------------

# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    callbacks=[early_stop]
)

# Predict on validation data
val_steps = validation_generator.samples // validation_generator.batch_size
predictions = model.predict(validation_generator, steps=val_steps, verbose=1)
predicted_classes = np.argmax(predictions, axis=1)

# True labels
true_classes = validation_generator.classes[:len(predicted_classes)]
class_labels = list(validation_generator.class_indices.keys())

# Classification Report
report = classification_report(true_classes, predicted_classes, target_names=class_labels)
print(report)

# Overall Accuracy
accuracy = accuracy_score(true_classes, predicted_classes)
print(f"Validation Accuracy: {accuracy*100:.2f}%")

# -----------------------------
# Part 4: Visualizations
# -----------------------------

# Plotting accuracy and loss
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plotting loss
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Confusion Matrix
cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# -----------------------------
# Part 5: Comparison with Manual Feature Engineering
# -----------------------------

# Feature Extraction Using HOG
def extract_hog_features(image_path):
    """
    Extract HOG features from an image.
    """
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    image_resized = cv2.resize(image, (128, 128))
    features, hog_image = hog(image_resized, 
                              orientations=9, 
                              pixels_per_cell=(8, 8),
                              cells_per_block=(2, 2), 
                              visualize=True, 
                              multichannel=False)
    return features

# Prepare HOG dataset
hog_features = []
hog_labels = []
for label, class_dir in enumerate(os.listdir(OUTPUT_DIR)):
    class_path = os.path.join(OUTPUT_DIR, class_dir)
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        features = extract_hog_features(img_path)
        hog_features.append(features)
        hog_labels.append(label)

hog_features = np.array(hog_features)
hog_labels = np.array(hog_labels)

# Feature Extraction Using SIFT
def extract_sift_features(image_path):
    """
    Extract SIFT features from an image and return a fixed-length descriptor.
    """
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    image_resized = cv2.resize(image, (128, 128))
    sift = cv2.SIFT_create()
    keypoints, descriptors = sift.detectAndCompute(image_resized, None)
    if descriptors is not None:
        # Use the mean of descriptors
        return descriptors.mean(axis=0)
    else:
        # If no descriptors are found, return a zero array
        return np.zeros(128)

# Prepare SIFT dataset
sift_features = []
sift_labels = []
for label, class_dir in enumerate(os.listdir(OUTPUT_DIR)):
    class_path = os.path.join(OUTPUT_DIR, class_dir)
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        features = extract_sift_features(img_path)
        sift_features.append(features)
        sift_labels.append(label)

sift_features = np.array(sift_features)
sift_labels = np.array(sift_labels)

# Training ANN on HOG features
X_train_hog, X_test_hog, y_train_hog, y_test_hog = train_test_split(
    hog_features, hog_labels, test_size=0.2, random_state=42)

ann_hog = MLPClassifier(hidden_layer_sizes=(256,), max_iter=500, random_state=42)
ann_hog.fit(X_train_hog, y_train_hog)
pred_hog = ann_hog.predict(X_test_hog)
print("HOG Feature ANN Classification Report:")
print(classification_report(y_test_hog, pred_hog))
print(f"HOG Feature ANN Accuracy: {accuracy_score(y_test_hog, pred_hog)*100:.2f}%")

# Training ANN on SIFT features
X_train_sift, X_test_sift, y_train_sift, y_test_sift = train_test_split(
    sift_features, sift_labels, test_size=0.2, random_state=42)

ann_sift = MLPClassifier(hidden_layer_sizes=(256,), max_iter=500, random_state=42)
ann_sift.fit(X_train_sift, y_train_sift)
pred_sift = ann_sift.predict(X_test_sift)
print("SIFT Feature ANN Classification Report:")
print(classification_report(y_test_sift, pred_sift))
print(f"SIFT Feature ANN Accuracy: {accuracy_score(y_test_sift, pred_sift)*100:.2f}%")

# Comparing Performance
print("\nPerformance Comparison:")
print(f"CNN Validation Accuracy: {accuracy*100:.2f}%")
print(f"HOG Feature ANN Accuracy: {accuracy_score(y_test_hog, pred_hog)*100:.2f}%")
print(f"SIFT Feature ANN Accuracy: {accuracy_score(y_test_sift, pred_sift)*100:.2f}%")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "output_black_white_cleaned_further1.jpg"  # Replace with your image path
image_cv = cv2.imread(image_path)

# Convert to grayscale
gray_image = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)

# Apply adaptive threshold to get binary image (invert colors to get black background)
binary_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 15, 10)

# Remove horizontal lines using morphological operations
horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (50, 1))  # Adjust the kernel size based on grid size
detect_horizontal = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

# Remove vertical lines using morphological operations
vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 50))  # Adjust the kernel size based on grid size
detect_vertical = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

# Combine the horizontal and vertical lines for the mask
grid_mask = cv2.add(detect_horizontal, detect_vertical)

# Refine the mask with additional morphological operations to remove dots and smaller lines
kernel = np.ones((3, 3), np.uint8)
grid_mask = cv2.dilate(grid_mask, kernel, iterations=1)
grid_mask = cv2.erode(grid_mask, kernel, iterations=1)

# Inpainting to remove the grid lines and fill them with surrounding pixel values
inpainted_image = cv2.inpaint(image_cv, grid_mask, 5, cv2.INPAINT_TELEA)

# Convert the inpainted image to grayscale again for black and white conversion
gray_inpainted = cv2.cvtColor(inpainted_image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred_image = cv2.GaussianBlur(gray_inpainted, (5, 5), 0)

# Apply median filtering to further remove salt-and-pepper noise
median_filtered_image = cv2.medianBlur(blurred_image, 5)

# Apply binary threshold to convert to a black and white image
_, black_and_white_image = cv2.threshold(median_filtered_image, 127, 255, cv2.THRESH_BINARY)

# Apply morphological opening to further clean up noise (removes small dots)
clean_kernel = np.ones((2, 2), np.uint8)
final_cleaned_image = cv2.morphologyEx(black_and_white_image, cv2.MORPH_OPEN, clean_kernel, iterations=2)

# Save the final processed image
final_image_path = "output_black_white_cleaned_further2.jpg"
cv2.imwrite(final_image_path, final_cleaned_image)

# Display the final image
plt.imshow(final_cleaned_image, cmap="gray")
plt.axis("off")  # Hide axes
plt.show()
```

```python
import cv2
import numpy as np
import os

# Load the image
image_path = "output_black_white_cleaned_further2.jpg"  # Replace with your image path
image_cv = cv2.imread(image_path)

# Convert to grayscale
gray_image = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)

# Apply adaptive threshold to get binary image (invert colors to get black background)
binary_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 15, 10)

# Remove noise using GaussianBlur
blurred_image = cv2.GaussianBlur(binary_image, (5, 5), 0)

# Find contours (objects) in the image
contours, _ = cv2.findContours(blurred_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a directory to save the detected objects
output_dir = "detected_objects"
os.makedirs(output_dir, exist_ok=True)

# Set the minimum area threshold for objects
min_area = 2000  # You can adjust this value to exclude smaller objects

# Iterate over each detected contour
object_counter = 1
for contour in contours:
    # Get the bounding box for the contour
    x, y, w, h = cv2.boundingRect(contour)
    area = w * h  # Calculate the area of the bounding box

    # Filter out small objects based on the area
    if area > min_area:  # Only process objects with an area larger than the threshold
        # Crop the object from the original image
        cropped_object = image_cv[y:y+h, x:x+w]

        # Save the detected object
        object_path = os.path.join(output_dir, f"object_{object_counter}.jpg")
        cv2.imwrite(object_path, cropped_object)

        # Increment the counter for object files
        object_counter += 1

# Optional: draw contours and bounding boxes on the original image for visualization
image_with_contours = image_cv.copy()
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    area = w * h
    if area > min_area:  # Same area threshold as above
        cv2.rectangle(image_with_contours, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Save the image with contours drawn (optional)
cv2.imwrite("image_with_contours_large_objects.jpg", image_with_contours)

# Display the image with contours drawn
plt.imshow(cv2.cvtColor(image_with_contours, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
